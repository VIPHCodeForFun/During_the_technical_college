Übung 5:
Generiere eine Menge an 1000 zufälligen Körpergrößen. Von diesen 1000 sollen 500 Männer
repräsentieren und 500 Frauen.


```python
# a) Die Körpergrößen der Frauen sollen normalverteilt sein und definiert werden durch
#    μ = 165 und  = 9. Des Weiteren gilt es jede weibliche Körpergröße mit einer "flag =0" zu markieren.

frauen = np.random.normal(165, 9, size=(500, 2)) # E(x)=μ=165  V(x)=9 n=500 

for i in range(len(frauen)):
    frauen[i][1]= 0

print(frauen)
# print(len(f))
```

```python
# b) Die Körpergrößen der Männer sollen normalverteilt sein und definiert werden durch
#   μ = 175 und  = 10. Des Weiteren gilt es jede männliche Körpergröße mit einer "flag= 1" zu markieren.
#   Verwende die Variable f bzw m, um die Zufallszahlen zu speichern. Ergänze diese beiden
#   1-dimensionalen Matrizen um eine zweite Spalte welche die jeweilige flag-Information beinhalten
#   soll. Danach hänge die beiden 500x2 Matrizen zusammen und erzeuge mit Hilfe einer
#   geeigneten Funktion eine zufällige Reihenfolge der generierten Körpergrößen. Abschließend
#   speichere die Daten in einem File mit Namen bodyheight.dat.

mann = np.random.normal(175, 10, size=(500, 2)) # E(x)=μ=175  V(x)=10 n=500 

for i in range(len(mann)):
    mann[i][1]= 1

#print(mann)
sum = np.concatenate((frauen,mann))
print(sum)

np.random.shuffle(sum)
print(sum)

np.savetxt('GeilesTeil.txt', sum, fmt=['%f','%i']) #[float] [int]
```

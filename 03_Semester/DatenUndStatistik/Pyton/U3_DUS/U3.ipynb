dargestellt werden, dass das Ergebnis von 10a) - ˆm = 2 ¯m − 1 ein unbiased Schätzer für m -

```python
import numpy as np #lade numpy als np # muss nur ein mal erstellt werden

rng = np.random.default_rng()    #erzeuge das Element rng aus np
```

```python
import numpy as np
import matplotlib.pyplot as pp

rng = np.random.default_rng()
actualArray = np.arange(1, 101)
actualSize = len(actualArray)

def useSchaetzer(data, sampleSize):
    results = []
    for x in range(200):
        samples = rng.choice(a = data, size = sampleSize, replace=False)
        schaetzer = 2*np.mean(samples)-1
        results.append(schaetzer)
    return results

def displayData(results, actualSize, sampleSize):    
    pp.title("Samplesize = {}".format(sampleSize))
    text = "Erwartungswert: {:4.2f}\nStandardabweichung: {:4.2f}".format(np.average(results), np.std(results))
    pp.text(1, 0.7, text, fontsize=14, transform=pp.gcf().transFigure)

    pp.axvline(actualSize, c = 'r')    
    pp.plot(results, np.zeros_like(results), '|')
    pp.show()

results5 = useSchaetzer(actualArray, sampleSize = 5)
results10 = useSchaetzer(actualArray, sampleSize = 10)
results25 = useSchaetzer(actualArray, sampleSize = 25)
results50 = useSchaetzer(actualArray, sampleSize = 50)
results80 = useSchaetzer(actualArray, sampleSize = 80)

fig, ax = pp.subplots()
ax.boxplot([results5, results10, results25, results50, results80])
ax.set_xticklabels(["n=5", "n=10", "n=25", "n=50", "n=80"])
pp.show()

```

Übung 12 (P):
In einer Gruppe von Personen wählt jeder zufällig und unabhängig von den anderen 2 ”Freunde”
aus. X bezeichnet die Anzahl der Personen, die von niemandem gewählt werden.
Schätzen Sie den Erwartungswert von X durch eine geeignete Simulation.

```python
import math
rng = np.random.default_rng()

n_pers = 20
n_sims = 50
allFriends = np.arange(0, n_pers)
results = []

def choosePeopleNooneLikes():
    allChosenFriends = []
    friendsWhoNooneLikes = []
    for i in range(0, n_pers):
        possibleFriendsToChoose = [x for x in allFriends if x != i]
        chosenFriends = rng.choice(possibleFriendsToChoose, 2, replace=False)
        allChosenFriends.extend(chosenFriends)


    for friend in allFriends:
        isTheChosenOne = False
        for chosenFriend in allChosenFriends:
            if(friend == chosenFriend):
                isTheChosenOne = True
                break
        if(isTheChosenOne == False):
            friendsWhoNooneLikes.append(friend)
            
       # print("No one likes these people: {}".format(friendsWhoNooneLikes))
    return len(friendsWhoNooneLikes)

for x in range(0, n_sims):
    results.append(choosePeopleNooneLikes())
    
print("Erwartungswert: {}".format(math.floor(np.mean(results))))
```

Übung 15 (P):
Erzeugt Stichproben der Größe n 2 {100, 1000, 10000} aus einer Normalverteilung mit Erwartungswert
2 und Varianz 3. Berechne mit Hilfe von Python einen Punktschätzer für die
Erwartungswerte und Varianzen der Stichproben, sowie 95% Konfidenzintervalle für den Mittelwert.
Die Berechnung der KIs soll sowohl mit den geschätzten Varianzen als auch mit der
angegebenen Varianz 2 = 3 erfolgen.


```python
import numpy as np
from math import sqrt
from scipy import stats

rng = np.random.default_rng()
ns = [100, 1000, 10000]

for n in ns:
    muOriginal, sigmaOriginal = 2, sqrt(3) # mean and standard deviation
    normalVals = rng.normal(muOriginal, sigmaOriginal, n)
    muCalculated = np.average(normalVals)
    sigmaCalculated = np.std(normalVals)

    print("Für n = {}\nErwartungswert = {}\nVarianz = {}".format(n, muCalculated, np.var(normalVals)))
    
    standardErrorOriginal = sigmaOriginal / sqrt(n)
    standardErrorCalculated = stats.sem(normalVals)
    conf_int = stats.norm.interval(0.95, loc=muCalculated, scale=standardErrorCalculated)
    print( "Konfidenzintervall mit berechneten Werten = {}" .format(conf_int))
    
    conf_int = stats.norm.interval(0.95, loc=muCalculated, scale=standardErrorOriginal)
    print( "Konfidenzintervall mit originalen Werten = {}" .format(conf_int))
    print("\n")
```

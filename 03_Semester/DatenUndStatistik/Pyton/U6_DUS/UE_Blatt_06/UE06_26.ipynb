Lade den Datensatz corr. Berechne die Korrelationen (z.B. np.corrcoef) des ersten Spaltenvektors mit den restlichen Spaltenvektoren und stelle die Korrelationen graphisch dar (pyplot.scatter oder pyplot.matshow in Kombination mit pyplot.colorbar) und interpretiere das Ergebnis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

def printCorr(axs, x, y):
    m, b = np.polyfit(x, y, 1)
    np.corrcoef(x, y)
    axs.scatter(x, y, marker='o')
    axs.plot(x, m*x+b, c="red")

df = pd.read_csv('corr.txt', sep="\t", header=None)

n = 7
fig, axs = plt.subplots(n, 1, figsize=(10, 40), sharex=True, sharey=True)

for i in range(0, 7):
    printCorr(axs[i], df[0], df[i+1])

plt.show()
```

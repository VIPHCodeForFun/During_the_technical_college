Lade die Datei regression und berechne den Korrelationskoeffizienten zwischen den einzelnen Spalten. Gib für entsprechend hohe Korrelation die Regressionsgeraden an, welche sich entweder mit Hilfe der Parameter aus dem Skript oder mit Hilfe von Python-Libs berechnen lässt (hilfreich: https://databraineo.com/statistik/lineare-regression-in-python/). Interpretiere das Ergebnis. Welche Informationen kann durch die Berechnung der Regressionsgeraden ermittelt werden?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

def printCorr(axs, x, y):
    m, b = np.polyfit(x, y, 1)
    r = np.corrcoef(x, y)
    axs.scatter(x, y, marker='o')
    axs.plot(x, m*x+b, c="red")
    if abs(r[0, 1]) > 0.75:
        text = f"""
        correlating
        correlation: {r[0, 1]}
        """
    else:
        text = f"""
        not correlating
        correlation: {r[0, 1]}
        """
    plt.text(1, 0.3, text, fontsize=14, transform=axs.transAxes)

df = pd.read_excel(
    'regression.xlsx', 
    names=["koerpergroesse", "schuhgroesse", "notendurchschnitt", "gewicht", "einkommen"])

fig, axs = plt.subplots(10, 1, figsize=(10, 40))

printCorr(axs[0], df["koerpergroesse"], df["schuhgroesse"])
printCorr(axs[1], df["koerpergroesse"], df["notendurchschnitt"])
printCorr(axs[2], df["koerpergroesse"], df["gewicht"])
printCorr(axs[3], df["koerpergroesse"], df["einkommen"])
printCorr(axs[4], df["schuhgroesse"], df["notendurchschnitt"])
printCorr(axs[5], df["schuhgroesse"], df["gewicht"])
printCorr(axs[6], df["schuhgroesse"], df["einkommen"])
printCorr(axs[7], df["notendurchschnitt"], df["gewicht"])
printCorr(axs[8], df["notendurchschnitt"], df["einkommen"])
printCorr(axs[9], df["gewicht"], df["einkommen"])

plt.show()
```
